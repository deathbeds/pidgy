Link definitions and footnote definitions can be used to defined dictionaries or enrich a documents annotaitons.

```python
    q=\
[a]: This
[b]: That
    
    >>> q
    {'a': 'This', 'b': 'That'}

[c]: Them
[d]: They
    
    >>> __annotations__
    {'c': 'Them', 'd': 'They'}
```

```python
    >>> __annotations__
    {'f': 'That', 'e': 'This', 'c': 'Them', 'd': 'They'}
    
[^e]: This
[^f]: That
```

```python
    qq: "I'm the boss"
    >>> assert 'qq' not in locals()
    >>> __annotations__['qq']
    "I'm the boss"
    
[qq]: I'm
[gg]: asd
```
