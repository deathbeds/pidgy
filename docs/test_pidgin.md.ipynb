    def test_inline_failure(ip):
        with pytest.raises(pidgin.post_run_cell.InlineCodeException):
            result = ip.run_cell("""`assert False`""")
        assert type(result.error_in_exec) == pidgin.post_run_cell.InlineCodeException

    def test_doctest_failure(ip):
        result = ip.run_cell(""">>> assert False""")
        assert type(result.error_in_exec) == pidgin.post_run_cell.DocTestException

    def test_jinja_failure(ip):
        result = ip.run_cell("""{{range}""")
        assert isinstance(result.error_in_exec, jinja2.TemplateError)
